```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import csv
```

```python
url='https://finance.yahoo.com/quote/0700.HK/history?period1=1509292800&period2=1540828800&interval=1d&filter=history&frequency=1d'
r=requests.get('https://finance.yahoo.com/quote/0700.HK/history?period1=1509292800&period2=1540828800&interval=1d&filter=history&frequency=1d')
r.encoding='utf-8'
html_str=r.text
data=BeautifulSoup(html_str,"html.parser")
```

```python
a=data.find_all('td')
len(a)
list=[]
for i in range(0,len(a)):
    list.append(a[i].text.strip())
len(list)
list
```

```python
with open('Tencent.csv','w',newline='')as f:
    writer=csv.writer(f)
    header=['Date','Open','High','Low','Close','Adj Close','Volume']
    writer.writerow(header)
    for i in range(0,100):
        writer.writerow([list[i*7],list[i*7+1],list[i*7+2],list[i*7+3],list[i*7+4],list[i*7+5],list[i*7+6]])
```

```python
df=pd.read_csv('Tencent.csv')
df['Difference']=df['High']-df['Low']
del df['Adj Close']
df
#df['Month']=df['Difference']
#df
#df['Date'].replace({'Oct 30,2018','October'})
#df
#df['prices'].replace({'$101.00':'$101-200'}).value_counts()
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df['datetime'] = df['Date'].apply(parse_datetime)
df
```

```python
#news point 1: the max price was continueingly decreasing from June 2018 to October 2018
df.set_index('datetime').resample('M').aggregate('max').plot()
```

```python
#news point 2: stock price experienced violent change during August and October.
df[
    (df['datetime']>datetime(2018,6,1)) &
    (df['datetime']<datetime(2018,6,30))
].set_index('datetime').resample('D').aggregate('max').plot()
df[
    (df['datetime']>datetime(2018,7,1)) &
    (df['datetime']<datetime(2018,7,31))
].set_index('datetime').resample('D').aggregate('max').plot()
df[
    (df['datetime']>datetime(2018,8,1)) &
    (df['datetime']<datetime(2018,8,31))
].set_index('datetime').resample('D').aggregate('max').plot()
df[
    (df['datetime']>datetime(2018,9,1)) &
    (df['datetime']<datetime(2018,9,30))
].set_index('datetime').resample('D').aggregate('max').plot()
df[
    (df['datetime']>datetime(2018,10,1)) &
    (df['datetime']<datetime(2018,10,31))
].set_index('datetime').resample('D').aggregate('max').plot()
```

```python
#news point 2: open and close price did not vary too much
sns.regplot(df['Open'], df['Close'])
ax = sns.scatterplot(x="Open", y="Close",hue='Difference',data=df)
```
