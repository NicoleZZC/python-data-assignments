```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import csv
```

```python
url='https://finance.yahoo.com/quote/0700.HK/history?period1=1509292800&period2=1540828800&interval=1d&filter=history&frequency=1d'
r=requests.get('https://finance.yahoo.com/quote/0700.HK/history?period1=1509292800&period2=1540828800&interval=1d&filter=history&frequency=1d')
r.encoding='utf-8'
html_str=r.text
data=BeautifulSoup(html_str,"html.parser")
```

```python
a=data.find_all('td')
len(a)
list=[]
for i in range(0,len(a)):
    list.append(a[i].text.strip())
len(list)
list
```

```python
with open('Tencent.csv','w',newline='')as f:
    writer=csv.writer(f)
    header=['Date','Open','High','Low','Close','Adj Close','Volume']
    writer.writerow(header)
    for i in range(0,100):
        writer.writerow([list[i*7],list[i*7+1],list[i*7+2],list[i*7+3],list[i*7+4],list[i*7+5],list[i*7+6]])
```

```python
df=pd.read_csv('Tencent.csv')
df['Difference']=df['High']-df['Low']
del df['Adj Close']
df
```

```python
%matplotlib inline
ax=df['High'].hist(bins=10)
plt.title('Daily Highest Price')
```
